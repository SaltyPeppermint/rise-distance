# Guided-Search Grid

Compare sampling strategies and root-size distributions across repeated sampling seeds.
Each observation says whether a seed/goal pair was reached within the selected candidate-prefix budget.

```python
import altair as alt

import helpers as H
import plots as P

alt.theme.register("analysis", enable=True)(lambda: P.THEME)
alt.data_transformers.enable("default", max_rows=None)
```

```python
# Empty string selects the newest completed/partially completed grid.
# Use e.g. "run.2" to select a particular data/guided_search_grid/run.* folder.
GRID = H.resolve_grid("")

# Available cumulative prefixes were recorded by guided_search_grid.py.
BUDGETS = H.grid_prefix_budgets(GRID)
BUDGETS
```

```python
# Default to the largest recorded prefix; replace with another value from BUDGETS.
BUDGET = BUDGETS[-1]

df, meta = H.load_grid(GRID, BUDGET)
gr = H.grid_goal_reach(df)
```

## Reachability summary

Per-pair reach rate across sampling seeds, mean pair reachability, and coverage by mode.

```python
P.reachability(df, gr, meta)
```
